# Artwork KAPPA, Percentage of agreetment and Randolph score : 5 reviewers

```python
import pandas as pd

df = pd.read_csv("/Users/danielafe/Desktop/ISWSvOCANS/Artwork Numbers Fleiss Kappa Step by Step - Copia de RawData.csv")
df = df.drop('Rater 6 (Script)', axis=1)
df
```

```python
import numpy as np
import statsmodels.api as sm
from statsmodels.regression.mixed_linear_model import VCSpec
from statsmodels.stats.inter_rater import fleiss_kappa
from statsmodels.stats.inter_rater import aggregate_raters

import pandas as pd


list_agreetments = list()
for (index, column) in enumerate(df):
    #print (index, np.asarray(df[column]))
    for (index2, column_other) in enumerate(df):
        if (index != index2) and (index2  > index ):
            #print (index, np.asarray(df[column]))
            #print (index2, np.asarray(df[column_other]))
            _agrets = []
            for element1, element2 in  zip(df[column],df[column_other]):
                #print(element1, element2)
                if element1 == element2:
                    _agrets.append(1)
                else:
                    _agrets.append(0)
            #print(_agrets)
            list_agreetments.append(_agrets)
                                           
#    print("*"*50)
 

#for _agrets in list_agreetments:
#    print(_agrets)
#    print(sum(_agrets))

#print(list_agreetments)
import numpy as np
all_agrett = (np.array(list_agreetments) ).T
_mean = list()
for row in all_agrett:
    #print(row)
    value = sum(row) / len(row)
    _mean.append(value)

print("Percentage agretment: ",sum(_mean)/len(_mean))


giro = np.array(df.values).transpose()
giro = df.values#np.array(table) .transpose()
print("Randolph Kappa: ",fleiss_kappa(aggregate_raters(giro)[0], method='randolph')) # randolph fleiss
kappam_fleiss(df, detail=True)
```
